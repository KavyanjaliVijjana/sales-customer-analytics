```python
import pandas as pd


import matplotlib.pyplot as plt

df = pd.read_csv(
    r"C:\Users\anju&janu\OneDrive\kavya\sales-customer-analytics\data\superstore_sales.csv",
    encoding="cp1252"
)
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(8,5))
region_sales.plot(kind='bar')

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.savefig("../images/region_sales.png")
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Month'] = df['Order Date'].dt.month
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(kind='line', marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()

plt.savefig("../images/Monthly_Sales_Trend.png")
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(7,7))
category_sales.plot(kind='pie', autopct='%1.1f%%')

plt.title("Sales Distribution by Category")

plt.ylabel("")

plt.show()

plt.savefig("../images/Sales_Distribution_by_Category.png")
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Sales'], bins=30)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()

plt.savefig("../images/Sales_Distribution.png")
```

```python
top_products = df.groupby('Product Name')['Sales'].sum()

top_products = top_products.sort_values(ascending=False)

print(top_products.head(5))
```

```python
category_profit = df.groupby('Category')['Profit'].sum()

plt.figure(figsize=(8,5))

category_profit.plot(kind='bar')

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.show()
```
